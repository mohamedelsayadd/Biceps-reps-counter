# **Define a function to calculate the angle Between Corners**

```python
def calculate_angle(a,b,c) : 
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    radians = np.arctan2(c[1]-b[1],c[0]-b[0]) - np.arctan2(a[1]-b[1],a[0]-b[0])
    angle = np.abs(radians*(180/np.pi))

    if angle > 180 : 
        angle = 360 - angle 
    return angle  
```

# **here we can show the main code**

```python
# Open the video file for reading
cap = cv2.VideoCapture("me_gym.mp4")


frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Create a VideoWriter object to save the processed video
out = cv2.VideoWriter("rusult.avi", cv2.VideoWriter_fourcc(*"XVID"), fps, (frame_width, frame_height))

counter = 0  # Initialize counter for counting reps
stage = None  # Initialize stage variable to track exercise stage

while True:
    # Read a frame from the video
    ret, frame = cap.read()
    if not ret:
        break  # Exit loop if no more frames

    # Convert the frame to RGB format
    RGB_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    # Process the frame using the pose detection model
    result = pose.process(RGB_frame)
    
    # Convert the frame back to BGR format
    BGR_frame = cv2.cvtColor(RGB_frame, cv2.COLOR_RGB2BGR)

    try:
        # Extract landmarks from the pose detection result
        landmarks = result.pose_landmarks.landmark

        # Get the coordinates of the right shoulder, elbow, and wrist
        r_shoulder = [landmarks[mPose.PoseLandmark.RIGHT_SHOULDER.value].x, landmarks[mPose.PoseLandmark.RIGHT_SHOULDER.value].y]
        r_elbow = [landmarks[mPose.PoseLandmark.RIGHT_ELBOW.value].x, landmarks[mPose.PoseLandmark.RIGHT_ELBOW.value].y]
        r_wrist = [landmarks[mPose.PoseLandmark.RIGHT_WRIST.value].x, landmarks[mPose.PoseLandmark.RIGHT_WRIST.value].y]
        
        # Calculate the angle of the right arm
        angle = calculate_angle(r_shoulder, r_elbow, r_wrist)  

        # Define font for text display
        font = cv2.FONT_HERSHEY_TRIPLEX
        
        # Display the calculated angle on the frame
        cv2.putText(BGR_frame, str(angle), tuple(np.multiply(r_elbow, [frame_width , frame_height]).astype(int)), font, 1, (255,0,0), 3)

        # Determine the exercise stage based on the angle
        if angle > 140:
            stage = "Down"
        elif angle < 75 and stage == "Down":
            stage = "Up"
            counter += 1  # Increment counter when the stage changes
            print(counter)  # Print counter value

    except Exception as e:
        print(e)  # Print exception if needed for debugging

    # Draw a white rectangle on the frame for displaying stats
    cv2.rectangle(BGR_frame, (0, 0), (300, 200), (255, 255, 255), -1)

     # Define font for text display
    font = cv2.FONT_HERSHEY_TRIPLEX
    
    # Display the rep counter and current stage on the frame
    cv2.putText(BGR_frame, "Reps", (15, 50), font, 1, (255, 0, 0), 2)
    cv2.putText(BGR_frame, str(counter), (15, 150), font, 2, (255, 0, 0), 2)
    cv2.putText(BGR_frame, "Stage", (120, 50), font, 1, (255, 0, 0), 2)
    cv2.putText(BGR_frame, stage, (120, 150), font, 1, (255, 0, 0), 2)

    # Draw pose landmarks and connections on the frame
    drawing.draw_landmarks(BGR_frame, result.pose_landmarks, mPose.POSE_CONNECTIONS)

    # Write the processed frame to the output video file
    out.write(BGR_frame)
    
    # Resize the frame for display
    BGR_frame = cv2.resize(BGR_frame, (480 , 720))
    
    # Display the frame in a window
    cv2.imshow("Biceps Counter", BGR_frame)
    
    # Exit loop if the 'Esc' key is pressed
    if cv2.waitKey(10) == 27:
        break

# Release video capture and writer objects
cap.release()
out.release()

# Close all OpenCV windows
cv2.destroyAllWindows()
```
